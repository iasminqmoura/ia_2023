```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tempo_servico = np.random.uniform(0, 10, 50)
salario = 1500 + 300 * tempo_servico + np.random.normal(0, 200, 50)
# salario = np.random.uniform(0, 2000, 50)
```

```python
dados = {
    'Tempo de serviço' : tempo_servico,
    'Salário' : salario
}
df = pd.DataFrame(dados)
df
```

```python
df.corr()
```

```python
# Cálculo da regressão linear
mean_x = np.mean(tempo_servico) # média do x (tempo de serviço)
mean_y = np.mean(salario) # média do y (salário)

numerador = np.sum((salario - mean_y) * (tempo_servico - mean_x))
denominador = np.sum((tempo_servico - mean_x) ** 2)
coeficiente = numerador/denominador

intercept = mean_y - coeficiente * mean_x
y_pred = intercept + coeficiente * tempo_servico
```

```python
# Aplicação prática
novo_tempo = 8.5
novo_salario = coeficiente * novo_tempo + intercept
novo_salario
```

```python
# Plotar gráfico
plt.scatter(tempo_servico, salario, color='blue', label='Dados aleatorios')
plt.plot(tempo_servico, y_pred, color='red', label='Regressão linear')
plt.scatter(novo_tempo, novo_salario, color='black', s=100)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título do gráfico')
plt.legend()
# plt.grid(True)
plt.show()
```
