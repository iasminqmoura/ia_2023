```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

valores1 = np.random.randint(0, 101, 50)

df1 = pd.DataFrame({'Valores': valores1})

plt.scatter(df1.index, df1['Valores'], marker='o', color='blue')

plt.xlabel('Amostra')
plt.ylabel('Valores')
plt.title('Gráfico de Valores Aleatórios')

plt.show()
```

```python
valores2 = np.random.uniform(-10, 10, 100)

df2 = pd.DataFrame({'Valores': valores2})

plt.scatter(df2.index, df2['Valores'], marker='D', color='yellow')

plt.xlabel('Amostra')
plt.ylabel('Valores')
plt.title('Gráfico de Valores Aleatórios Entre -10 e 10')

plt.show()
```

```python
np.random.seed(0)
tempo_atividade = np.random.uniform(0, 8, 50)
calorias_gastas = 300 + 100 * tempo_atividade + np.random.normal(0, 40, 50)

df3 = pd.DataFrame({'Tempo de Atividade (horas)': tempo_atividade, 'Calorias Gastas': calorias_gastas})

plt.scatter(df3['Tempo de Atividade (horas)'], df3['Calorias Gastas'], marker='o', color='green')

plt.xlabel('Tempo de Atividade (horas)')
plt.ylabel('Calorias Gastas')
plt.title('Gráfico de Dispersão: Tempo de Atividade x Calorias Gastas')

plt.show()
```

```python
df_escola = pd.read_csv('escola_y_dataset.csv')

idade = df_escola['Idade']
faltas = df_escola['Faltas']

regressao = linregress(idade, faltas)

plt.scatter(idade, faltas, label='Dados', color='pink')

plt.plot(idade, regressao.intercept + regressao.slope * idade, color='red', label='Reta de Regressão')

plt.xlabel('Idade')
plt.ylabel('Faltas')
plt.title('Regressão Linear: Idade x Faltas')

plt.legend()

plt.show()
```

```python
df_academia = pd.read_csv('academia_z_dataset.csv')

tempo_atividade = df_academia['Tempo_Atividade']
calorias = df_academia['Calorias']

regressao = linregress(tempo_atividade, calorias)

plt.scatter(tempo_atividade, calorias, label='Dados', color='purple')

plt.plot(tempo_atividade, regressao.intercept + regressao.slope * tempo_atividade, color='red', label='Reta de Regressão')

plt.xlabel('Tempo de Atividade (horas)')
plt.ylabel('Calorias Gastas')
plt.title('Regressão Linear: Tempo de Atividade x Calorias Gastas')

plt.legend()

plt.show()
```

```python
df_empresa = pd.read_csv('empresa_x_dataset.csv')

tempo_servico = df_empresa['Tempo_de_Servico']
salario = df_empresa['Salario']

regressao = linregress(tempo_servico, salario)

plt.scatter(tempo_servico, salario, label='Dados', color='orange')

plt.plot(tempo_servico, regressao.intercept + regressao.slope * tempo_servico, color='red', label='Reta de Regressão')

plt.xlabel('Tempo de Serviço (anos)')
plt.ylabel('Salário')
plt.title('Regressão Linear: Tempo de Serviço x Salário')

plt.legend()

plt.show()
```
